```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
```

```python
df_path = "C:\\Users\\HP\\Downloads\\Dataset MeriSKILL\\Sales Data.csv"
df = pd.read_csv(df_path)
df
```

```python
df.head(10)
```

# Total sales

```python
total_sales = df['Sales'].sum()
print('Total Sales : ', total_sales)
```

# Total sale for each product

```python
total_sales_for_each_product = df.groupby('Product')['Sales'].sum()
prod_sal = total_sales_for_each_product.reset_index(name = 'Total Sales')
prod_sal
```

# Highest product sale

```python
best_pro = print(total_sales_for_each_product.idxmax(), ' : ',total_sales.max() )
```

```python
sns.set_style('darkgrid')
sns.barplot(x = prod_sal.index, y = prod_sal['Total Sales'])
plt.title('Visualization of the best-selling products')
plt.ylabel('Total Sales')
plt.xlabel('Products')
plt.show()
```

# The total profit margins

```python
df['Profit'] = df['Sales'] - df['Price Each'] 
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
total_profit_margin = df['Profit Margin (%)'].sum()
print("\nTotal Profit Margins: {:.2f}%".format(total_profit_margin))
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='Product', y='Profit Margin (%)', data=df, palette='viridis')
plt.xlabel('Product')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margins by Product')
plt.ylim(0, 20)  
plt.xticks(rotation=45)
plt.tight_layout()
```
